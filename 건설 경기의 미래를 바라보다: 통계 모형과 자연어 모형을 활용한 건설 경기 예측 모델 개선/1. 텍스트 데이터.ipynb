```python
# matplotlib 한글 폰트 셋업
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
fm.fontManager.addfont(fontpath)  # 에전의 rebuild 대신 이걸 추가해야 설치한 폰트 추가됨
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic')
plt.rc('axes', unicode_minus=False)  # Glyph 8722 방지용
```

### 연도별 Word Cloud

```python
bow_2017 = pd.read_csv('../DATA_text/bow_2017.csv')
bow_2018 = pd.read_csv('../DATA_text/bow_2018.csv')
bow_2019 = pd.read_csv('../DATA_text/bow_2019.csv')
bow_2020 = pd.read_csv('../DATA_text/bow_2020.csv')
bow_2021 = pd.read_csv('../DATA_text/bow_2021.csv')
bow_2022 = pd.read_csv('../DATA_text/bow_2022.csv')
bow_2023 = pd.read_csv('../DATA_text/bow_2023.csv')
```

```python
stopwords_list = ['date','기사','뉴스','전자신문','최근','기자','구독','언론사','현재',
                 '로','를','것','및','말','등','이','대비','수','위','중','더','재','때','고','그','전','개','제',
                 '가장','기준','수준','지난','보기','경우','때문','대한','관련']
```

```python
# remove stopwords
bow_2017_stop = bow_2017.copy()

for col in stopwords_list:
    if col in bow_2017_stop.columns:
        bow_2017_stop.drop(col, axis=1, inplace=True)
```

```python
bow_2017_stop_sum = pd.DataFrame(bow_2017_stop.sum()).T
bow_2017_stop_sum = bow_2017_stop_sum[bow_2017_stop_sum.iloc[0].sort_values(ascending=False).index].iloc[:,:50]

# Gendrate WordCloud Object
wordcloud = WordCloud(font_path = fontpath, width=800, height=400, background_color='white')

# Word & Word bins to dict 
word_freq = bow_2017_stop_sum.to_dict(orient='records')[0]

# Gendrate WordCloud
wordcloud.generate_from_frequencies(word_freq)

# Draw WordCloud 
plt.figure(figsize=(5,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('2017년', fontsize = 15)
plt.tight_layout()
plt.savefig('../DATA_text_output/2017년 WordlCloud.png', dpi=300)
plt.show()
```

### 연도별 필터링 뉴스 갯수

```python
news_data = pd.read_csv('../DATA_text/filtered_news_lists_2000.csv')
news_data['date'] = news_data['date'].astype('str')
news_data['month'] = news_data['date'].apply(lambda x: x[:6])
news_data
```

```python
news_count = news_data.groupby('month').count()['news_id']
news_count = news_count.reset_index()
news_count = news_count.rename(columns = {'month':'기간','news_id':'뉴스 개수'})
news_count
```

```python
ax = news_count.set_index("기간").plot.bar(figsize=(10,5), width=0.8, color='darkgray')

# x축 ticker를 5단위로 설정
ax.xaxis.set_major_locator(MultipleLocator(base=10))
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.title('월별 뉴스 개수')
plt.grid(True)
plt.savefig('../DATA_text_output/월별 뉴스 개수.png', dpi=300)
plt.show()
```

```python
news_count["뉴스 개수"].mean()
```

### 연도별 키워드 등장 횟수

```python
news_count = news_data.groupby('month').sum()[['부동산','건설']]
news_count = news_count.reset_index()
news_count = news_count.rename(columns = {'month':'기간','부동산':'부동산 등장 횟수','건설':'건설 등장 횟수'})
news_count
```

```python
# 월별 뉴스 개수를 막대 그래프로 그리기
ax = news_count.set_index("기간").plot.bar(figsize=(10,5), width=0.8, color=['#A3C8FF', '#FF8D8D'])

# x축 ticker를 5단위로 설정
ax.xaxis.set_major_locator(MultipleLocator(base=10))
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.title('월별 키워드 등장 횟수')
plt.grid(alpha=0.3)
plt.savefig('../DATA_text_output/월별 키워드 등장 횟수.png', dpi=300)
plt.show()
```
