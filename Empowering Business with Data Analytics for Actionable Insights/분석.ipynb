# 2. 영업이익 관계 변수 파악

```python
import pandas as pd
region_category = pd.read_csv('./data/region_category_preprocess.csv',encoding='euc-kr')
region_category
```

```python
region_category.columns
```

### 1) 변수 간 상관관계 파악

```python
region_category.info()
```

```python
category_cols = region_category.select_dtypes(include=['object']).columns.tolist()
numeric_cols = region_category.select_dtypes(include=['number']).columns.tolist()
numeric_cols
```

```python
profit_col = ['영업이익총액','영업이익평균액','영업이익중위액']
for value in profit_col:
    if value in numeric_cols:
        numeric_cols.remove(value)
numeric_cols
```

```python
fig,axs = plt.subplots(figsize=(15,15),ncols=3,nrows=4)
for i,feature in enumerate(numeric_cols):
    row = int(i/3)
    col = i%3
    sns.regplot(x=feature,y='영업이익총액',data=region_category,ax = axs[row][col])
```

#### 2) 회귀분석을 통한 영업이익 예측 모델 제작

```python
from sklearn.model_selection import train_test_split
num_col = region_category.select_dtypes(include=['number']).columns.tolist()
X = region_category[num_col].drop(columns = '영업이익총액')
y = region_category['영업이익총액']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=156)
```

- 모델 학습

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error 
rf = RandomForestRegressor(random_state=0, n_estimators=70)
rf.fit(X_train,y_train)
y_preds = rf.predict(X_test)
```

- Feature Importance

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib as mpl
plt.rc('font', family='NanumBarunGothic')
mpl.rc('axes', unicode_minus=False)
```

```python
import seaborn as sns
%matplotlib inline
rf_reg = RandomForestRegressor(n_estimators=10)
rf_reg.fit(X_train,y_train)
feature_series = pd.Series(data=rf_reg.feature_importances_,index=X_train.columns)
feature_series = feature_series.sort_values(ascending=False)
sns.barplot(x=feature_series, y=feature_series.index)
```
